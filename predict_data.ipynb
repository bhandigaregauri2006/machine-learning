```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv('Salary_Data.csv')
ds
```

```python
#divded  data to depended

x=ds[['YearsExperience']]
y=ds[['Salary']]
```

```python
#scatter  plot
plt.scatter(x,y)
plt.show()

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test ,y_train,y_test=train_test_split(x,y,test_size=10,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
result=(reg.coef_*[10])+reg.intercept_
result
```

```python
result = reg.predict([[10]])
result
```

```python
result= reg.predict(x_test)
result
```

```python
y_test
```

```python
plt.scatter(x,y)
result=reg.predict(x)
plt.plot(x,result,color='Red')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
result = reg.predict(x_test)
print("MSE :",mean_squared_error(y_test,result))#actual real life values predict
print("R2 Score:",r2_score(y_test,result))
```
