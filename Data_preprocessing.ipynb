```python
#data cleaning -using scikit learn
from sklearn.impute import SimpleImputer
ds=pd.read_csv('data_preprocessing.csv')
ds
```

```python
si = SimpleImputer(missing_values=np.nan,strategy='mean')
ds[['Age','Salary']]=si.fit_transform(ds[['Age','Salary']])
ds
```

```python
#linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()

#simple data
x=[[1],[2],[3]]
y= [2,4,6]

#Training the model

model.fit(x,y)

#Predicting

print(model.predict([[10]]))
```

```python
#onehotencoder 
#example
#how to
from sklearn.preprocessing import OneHotEncoder 

#cerate object
ohe = OneHotEncoder(sparse_output=False)

#encode country column

encoded_data= ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
#get column
column_names = ohe.get_feature_names_out(['Country'])

#create new dataframe with encoded data and column name
encoded_df=pd.DataFrame(encoded_data,columns=column_names)

#concat
ds= pd.concat([ds,encoded_df],axis=1)


ds= ds.drop(columns=["Country"])
ds
```

```python
#labelEncoder encode target  labels with between 0 and 1_classes-1

from sklearn.preprocessing  import LabelEncoder
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
ds
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
ds[['Age','Salary']]=sc.fit_transform(ds[['Age','Salary']])
ds
```

```python
#ordinalEncoder 
from sklearn.preprocessing  import OrdinalEncoder

#sample data : categorical features with an inherent order
data = [['Medium'],['High'],['Low'],['Medium'],['High'],['Medium'],['Low'],['Low']]

#Initialize the ordinalencoder
encoder = OrdinalEncoder(categories=[['High','Medium','Low']]) #Define the order explicitly

#fit and transform the data
encoded_data = encoder.fit_transform(data)

#Display the encoded data
print("Original Data:",data)
print("Encoded Data :\n",encoded_data)
```
