```python
import matplotlib.pyplot as plt
x = ['a','b','c','d','e']
y = [7,34,56,12,78]
plt.figure(figsize = (8,5))
plt.plot(x,y,color = 'green',marker ='o') 
plt.title ("line chart")
plt.xlabel("Name of people")
plt.ylabel("Age of people")
for i in range(len(y)):
    plt.text(x[i],y[i]+1,y[i],fontsize=12,ha ='center')
plt.show()    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Values': [10, 10, 15]
})

# Set style and color palette
sns.set_style("ticks")  # Options: white, dark, whitegrid, darkgrid, ticks
sns.set_palette("colorblind")   # Try: deep, muted, bright, pastel, dark, colorblind

# Create the plot
plt.figure(figsize=(8, 5))
barplot = sns.barplot(x='Category', y='Values', data=data)

# Add value labels on top of bars
for p in barplot.patches:
    barplot.annotate(f'{p.get_height()}', 
                     (p.get_x() + p.get_width() / 2, p.get_height()), 
                     ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

# Customize title and labels
plt.title("Bar Plot of Categories", fontsize=14, fontweight='bold', color='navy')
plt.xlabel("Category", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Optional: Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optional: Remove top and right borders
sns.despine()

#count plot-A count plot is a type of bar plot that show the count of obs cat
# sns.countplot(x='category',data = data)
# plt.title("count plot")
# plt.show()
# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C', 'B', 'B']
})

# Plot
plt.figure(figsize=(7, 5))
sns.set_style("whitegrid")
sns.countplot(x='Category', data=data, palette='pastel')

plt.title("Count Plot of Categories", fontsize=14, fontweight='bold')
plt.xlabel("Category")
plt.ylabel("Count")

plt.show()
```

```python
#histogram using matplot
import numpy as np
data = np.random.randn(200) 
plt.hist(data, bins=10) 
plt.title("Histogram") 
plt.show() 
```

```python
#histogram using seaborn
sns.histplot(data, bins=20, kde=True) 
plt.title("Histogram with KDE") 
plt.show() 
```

```python
#kde plot
sns.kdeplot(data, shade=True) 
plt.title("KDE Plot") 
plt.show() 
```

```python
#box plot
sns.boxplot(y=data) 
plt.title("Box Plot") 
plt.show()
```

```python
#violin plot
sns.violinplot(y=data) 
plt.title("Violin Plot") 
plt.show()
```
